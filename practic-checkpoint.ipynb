```python
import numpy as np
import pandas as pd
```

```python
import glob
```

```python
folder_names = glob.glob("training_dataset\\*")
label = []
data = []
```

```python
print(folder_names)
```

```python
from PIL import Image 
```

```python
for folder in folder_names:
    filenames = glob.glob(folder+'\\*')
    
    for file in filenames:
        
        label = label+[[folder.split('\\')[1]]]
        img = Image.open(file).convert('RGB')
        img = img.resize((28,28))     
        img1 = np.asarray(img)   
        img1 = img1.reshape(-1,)
        
        data.append(img1)
        
```

```python
df1 = pd.DataFrame(data,columns=[str(i) for i in range(28*28*3)])
df2 = pd.DataFrame(label,columns=['label'])

df = df1.join(df2)
```

```python
df.to_excel('C:\\Users\\User\\Desktop\\colordata.xlsx')
```
